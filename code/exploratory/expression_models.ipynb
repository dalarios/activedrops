```python
import numpy as np
import matplotlib.pyplot as plt

# Define time range
t = np.linspace(0, 200, 1000)

# Define the protein production models
def baseline_model(t, kappa, gamma_m, D):
    return (kappa / gamma_m**2) * D * (np.exp(-gamma_m * t) - 1 + gamma_m * t)

def binding_model(t, kappa, gamma_m, D, K_d):
    return (kappa / gamma_m**2) * (D / (D + K_d)) * (np.exp(-gamma_m * t) - 1 + gamma_m * t)

def inhibition_model(t, kappa, gamma_m, D, K_d, K_i):
    return (kappa / gamma_m**2) * (D / (D + K_d + D**2 / K_i)) * (np.exp(-gamma_m * t) - 1 + gamma_m * t)

# Function to update plot
def update_plot(kappa=0.017, gamma_m=0.015, D=80.0, K_d=52.0, K_i=24.0):
    p_baseline = baseline_model(t, kappa, gamma_m, D)
    p_binding = binding_model(t, kappa, gamma_m, D, K_d)
    p_inhibition = inhibition_model(t, kappa, gamma_m, D, K_d, K_i)
    
    plt.figure(figsize=(10, 6))
    # plt.plot(t, p_baseline, label="Baseline Model")
    plt.plot(t, p_binding, label="Binding-Adjusted Model")
    plt.plot(t, p_inhibition, label="Inhibition-Adjusted Model")
    plt.xlabel("Time (min)")
    plt.ylabel("Protein Production (p(t))")
    plt.title("Protein Production Models")
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function with the specified parameters
update_plot()
```
