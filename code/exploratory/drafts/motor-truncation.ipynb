```python
# Define the file path
file_path = "C:/Users/dalar/Downloads/naegleria.txt"  # <-- Adjust this path
```

```python
import matplotlib.pyplot as plt

# Define the file path

# Read the file content
with open(file_path, 'r') as file:
    file_content = file.readlines()

# Extracting relevant data
positions = []
amino_acids = []
probabilities = []

# Parsing the relevant data lines (ignoring header lines and extracting relevant info)
for line in file_content[4:]:
    # Splitting the line into words
    words = line.split()
    
    # Extracting position, amino acid, and probability
    position = int(words[0])
    amino_acid = words[1]
    probability = float(words[3])
    
    # Appending to lists
    positions.append(position)
    amino_acids.append(amino_acid)
    probabilities.append(probability)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(positions, probabilities, marker='o', linestyle='-')
plt.title('Probability of Coiled Coil Formation Along Amino Acid Sequence')
plt.xlabel('Amino Acid Position')
plt.ylabel('Probability')
plt.xticks(positions[::20], rotation=45)  # Adjusted from 10 to 20 to reduce x-axis crowding
plt.grid(True, linestyle='--', alpha=0.5)

# Save the plot
output_file_path = "C:/Users/dalar/Downloads/naegleria.jpg"  # <-- Adjust this path
plt.tight_layout()
plt.savefig(output_file_path, format='jpg', dpi=300)

# Displaying the plot
plt.show()
```
