```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the matrix, treating the first row as data and assigning integer column names
df = pd.read_csv(
    "../../../../Thomson Lab Dropbox/David Larios/activedrops/chimeras-ADP-APO/APO/a/dccm_avg.csv",
    header=None
)
df.columns = range(df.shape[1])

# Flip the matrix vertically to correct the orientation
df_flipped = df.iloc[::-1].reset_index(drop=True)

# Define segment boundaries and labels
boundaries = [0, 401, 802, 1253, 1698]  # [start, K1, K2, alpha, beta, end]
labels = [
    "Kinesin 1", 
    "Kinesin 2", 
    "Alpha Tubulin", 
    "Beta Tubulin"
]

# Calculate tick positions (center of each segment)
tick_positions = [
    (boundaries[i] + boundaries[i+1]) // 2 for i in range(len(boundaries)-1)
]

# Adjust y-tick positions for the flipped matrix
n_rows = df.shape[0]
ytick_positions_flipped = [n_rows - 1 - pos for pos in tick_positions]

# Display the heatmap with labeled axes
plt.figure(figsize=(12, 10))
ax = sns.heatmap(df_flipped, cmap='coolwarm', cbar_kws={'label': 'DCCM value'})

# Set ticks and labels
ax.set_xticks(tick_positions)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=12)
ax.set_yticks(ytick_positions_flipped)
ax.set_yticklabels(labels, rotation=0, fontsize=12)

# Draw lines to separate the regions (adjust for flipped y-axis)
for b in boundaries[1:-1]:
    ax.axhline(n_rows - b, color='black', linewidth=1)
    ax.axvline(b, color='black', linewidth=1)

plt.title("Dynamic Cross Correlation Matrix\nKinesin Dimer with Alpha/Beta Tubulin", fontsize=16)
plt.xlabel("Residue Index / Region")
plt.ylabel("Residue Index / Region")
plt.tight_layout()
plt.show()
```
