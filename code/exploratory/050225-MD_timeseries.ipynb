```python
df = pd.read_csv("../../../../Downloads/k401_combined_distance.csv")

# Extract column names and sort them based on the first number in the name, except for the first column
sorted_columns = [df.columns[0]] + sorted(df.columns[1:], key=lambda x: int(x.split('_')[0]))

# Reorder the DataFrame columns
df = df[sorted_columns]
df
```

```python
import os, re
import numpy as np
import matplotlib.pyplot as plt

aa_classification = {
    'A': 'nonpolar', 'R': 'positively charged', 'N': 'polar uncharged', 'D': 'negatively charged', 'C': 'polar uncharged', 
    'Q': 'polar uncharged', 'E': 'negatively charged', 'G': 'nonpolar', 'H': 'positively charged', 'I': 'nonpolar', 
    'L': 'nonpolar', 'K': 'positively charged', 'M': 'nonpolar', 'F': 'nonpolar', 'P': 'nonpolar', 'S': 'polar uncharged', 
    'T': 'polar uncharged', 'W': 'nonpolar', 'Y': 'polar uncharged', 'V': 'nonpolar'
}

# Keep thread count low for the execution environment
os.environ["OPENBLAS_NUM_THREADS"] = "1"
os.environ["OMP_NUM_THREADS"] = "1"

# ---------- 1. Load the CLUSTAL alignment ------------------
aln_path = "../../../../Downloads/clustalo-I20250505-165631-0253-52195837-p1m.aln-clustal_num"

seq_chunks = {}  # sequence name â†’ list of alignment chunks
with open(aln_path) as fh:
    for line in fh:
        line = line.rstrip("\n")
        if (
            not line
            or line.startswith("CLUSTAL")
            or line.startswith("MUSCLE")
            or line.startswith(" ")
        ):
            continue

        m = re.match(r"(\S+)\s+([A-Za-z\-]+)", line)
        if m:
            name, chunk = m.groups()
            seq_chunks.setdefault(name, []).append(chunk)

# Join chunks into full alignment strings
align = {name: "".join(chunks) for name, chunks in seq_chunks.items()}

names = list(align.keys())
aln_len = len(next(iter(align.values())))
assert all(len(s) == aln_len for s in align.values())

# ---------- 2. Encode amino acids as integers --------------
# Simplified classification based on aa_classification
classification_list = sorted(set(aa_classification.values()))
classification_to_int = {classification: i for i, classification in enumerate(classification_list)}

matrix = np.zeros((len(names), aln_len), dtype=int)
for i, name in enumerate(names):
    seq = align[name]
    matrix[i] = [
        classification_to_int[aa_classification[aa]] if aa in aa_classification else -1
        for aa in seq
    ]

# ---------- 3. Plot heatmap --------------------------------
fig, ax = plt.subplots(figsize=(aln_len / 30 + 2, len(names) * 0.4 + 2))  # dynamic size
im = ax.imshow(matrix, aspect='auto', cmap='viridis')  # Use a colormap suitable for categorical data

# Yâ€‘axis: sequence names
ax.set_yticks(np.arange(len(names)))
ax.set_yticklabels(names)

# Xâ€‘axis: alignment positions (label every 50 residues for readability)
xticks = np.arange(0, aln_len, 50)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks + 1, rotation=90)

ax.set_title("Full Multiple Alignment (simplified classification)")

# Colorbar with classification legend
cbar = fig.colorbar(im, ax=ax, fraction=0.025, pad=0.02)
cbar.set_ticks(np.arange(len(classification_list)))
cbar.set_ticklabels(classification_list)
cbar.ax.set_ylabel("Classification")

plt.tight_layout()
plt.show()
```
