```python
## Import relevant libraries
import sys
sys.path.append('../../activedrops')

## Including ourselves
import quadrop as qd
qd.set_plotting_style()
```

```python
data_path = "../../data/020524-kif3-stirring/"
conditions = ["DNA40nM_MT", "DNA80nM_MT", "DNA40nM_NoMT", "DNA80nM_NoMT"]
subconditions = ["Pos0", "Pos1", "Pos2"]

# frame info
min_frame = 0
max_frame = None
skip_frames=8
frame_rate=30
time_intervals = [30, 30, 30, 30]


```

```python
# channel info
channel = "cy5"
vmax = 0.2

# Call the function with the updated parameters
qd.plot_fluorescence_vs_time(
    data_path=data_path,
    conditions=conditions,
    subconditions=subconditions,
    channel=channel,
    time_intervals=time_intervals,  
    min_frame=min_frame,
    max_frame=max_frame,
    skip_frames=skip_frames,
    log_scale=False,  # Set log_scale to True
    timescale="min"
)
```

```python
qd.single_fluorescence_movies(data_path, conditions, subconditions, channel, time_intervals, min_frame, max_frame, vmax, skip_frames, frame_rate)
```
